```python
from sklearn import preprocessing
obj = train.columns[train.dtypes==object]
for c in obj:
    lbl = preprocessing.LabelEncoder() 
    lbl.fit(list(train[c].values)) 
    print(c)
    test[c] = lbl.transform(list(test[c].values))
    train[c] = lbl.transform(list(train[c].values))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error,make_scorer
```

```python
parameters = {'alpha':np.linspace(0,10)}
```

```python
ridge = Ridge()
clf = GridSearchCV(ridge, parameters, cv=3,scoring=make_scorer(mean_squared_error))
clf.fit(train[train.columns[1:-1]],train[train.columns[-1]])
```

```python
mean_squared_error(clf.predict(train[train.columns[1:-1]]),train[train.columns[-1]])
```
